### Gradient Descent

```python
def J(X,y,theta):
    n = len(y)
    j = (np.dot(theta, X.T)-y)**2/(2*n)
    return j.sum()

def gradientDescent(X,y,its=1000, alpha=0.01):
    n = len(y)
    theta = np.array([0.,0.])
    Js= np.zeros(its)
    for it in range(its):
        theta -= np.dot(X.T,np.dot(theta,X.T)-y)*alpha/n
        Js[it] = J(X,y,theta)
    return theta, Js
```

```python
x = data[:,0]

# X now a matrix, including 1s (offset)
X = np.ones(shape=(data.shape[0],2))
X[:,1] = x
y = data[:,1]

theta, Js = gradientDescent(X,y, its=1500)

plt.figure()
plt.plot(x,y, ls='', marker='o', markersize=4)
plt.plot(x,np.dot(theta,X.T))
plt.xlabel("Population")
plt.ylabel("Profit");
print(f"Model: {theta[0]:.2f}+{theta[1]:.2f}x")

plt.figure()
plt.plot(Js)
plt.xlabel("Its")
plt.ylabel("J");
```

### Visualizing J($\theta$)

```python
def J_cont(X,y,tx,ty):
    n = len(y)
    print(tx.shape)
    js = np.zeros((n,n), dtype='float')
    for i in range(n):
        for j in range(n):
            theta = [tx[i][j], ty[i][j]]
            js[i,j] = np.sum((np.dot(theta, X.T)-y)**2)/(2*n)
    return js
```

```python
th0 = np.linspace(-7,2,len(y))
th1 = np.linspace(-3,5,len(y))
thx, thy = np.meshgrid(th0, th1)

plt.contour(thx, thy, J_cont(X,y,thx,thy),levels=100)
plt.plot(-3.63,1.17, ls='', marker='o', markersize=5)
```

### Linear regression with multiple variables

```python
data2 = np.genfromtxt('data2.txt', delimiter=',')
data2.shape
```

```python
def gradientDescent(X,y,its=1000, alpha=0.01):
    n = len(y)
    theta = np.array([0.,0.,0.])
    Js= np.zeros(its)
    for it in range(its):
        theta -= np.dot(X.T,np.dot(theta,X.T)-y)*alpha/n
        Js[it] = J(X,y,theta)
    return theta, Js
```

```python
x = data2[:,[0,1]]
X = np.ones_like(data2)
X[:,[1,2]]=data2[:,[0,1]]
y = data2[:,2]

X[:,1] = (X[:,1]-X[:,1].mean(axis=0))/np.std(X[:,1],ddof=1, axis=0)
X[:,2] = (X[:,2]-X[:,2].mean(axis=0))/np.std(X[:,2],ddof=1, axis=0)
# y = (y-y.mean())/np.var(y)

theta, Js = gradientDescent(X,y,its=400,alpha=0.01)
print(f"Model: {theta[0]:.0f} + {theta[1]:.0f}x1 + {theta[2]:.0f}x2")
```
